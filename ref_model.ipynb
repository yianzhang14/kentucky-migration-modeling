```python
df = pd.read_csv("data/psam_p21.csv")
df
```

```python
df["MIGPUMA"] = df["MIGPUMA"].fillna(0)
df["MIGSP"] = df["MIGSP"].fillna(0)
df["MIGPUMA"] = df["MIGSP"].astype(int).astype(str).str.zfill(2) + df["MIGPUMA"].astype(int).astype(str).str.zfill(5)
df["PUMA"] = df["ST"].astype(str).str.zfill(2) + df["PUMA"].astype(str).str.zfill(5)
df[["MIGPUMA", "PUMA"]]
```

```python
df[df["AGEP"] >= 18]
```

```python
# all migrations within Kentucky to Kentucky
df = df[df["AGEP"] >= 18]
df = df[(df["MIGPUMA"].isin(list(distances.index.astype(str))))]
df
```

```python
df["PUMA"] = df["PUMA"] + "1"
df["MIGPUMA"] = df["MIGPUMA"] + "2"
```

```python
df["filler"] = 0
```

```python
puma_migpuma = pd.read_excel("data/puma_migpuma.xlsx", dtype="str")
puma_migpuma["PUMA"] = puma_migpuma["State"] + puma_migpuma["PUMA"] + "1"
puma_migpuma["MIGPUMA"] = puma_migpuma["State"] + puma_migpuma["MIGPUMA"] + "2"
puma_migpuma = puma_migpuma.set_index("PUMA")
puma_migpuma.head()
```

```python
migration_matrix = df.groupby(["MIGPUMA", "PUMA"])["filler"].count().reset_index()
migration_matrix
```

```python
distances.index = distances.index + "2"
distances.columns = distances.columns + "1"
```

```python
migration_matrix["distance"] = 0
for i, row in migration_matrix.iterrows():
    migration_matrix.loc[i, "distance"] = distances.loc[row["MIGPUMA"], row["PUMA"]]
```

```python
migration_matrix["year"] = 2018
migration_matrix["internal"] = np.where(migration_matrix["distance"] == 0, 1, 0)
```

```python
migration_matrix["pop_orig"] = pops.loc[list(migration_matrix["MIGPUMA"].values), "Total Population"].values
migration_matrix["pop_dest"] = pops.loc[list(migration_matrix["PUMA"].values), "Total Population"].values
```

```python
migration_matrix["dens_orig"] = pops.loc[list(migration_matrix["MIGPUMA"].values), "Population Density (Per Sq. Mile)"].values
migration_matrix["dens_dest"] = pops.loc[list(migration_matrix["PUMA"].values), "Population Density (Per Sq. Mile)"].values
```

```python
migration_matrix
```

```python
migration_matrix = pd.merge(migration_matrix, migration_matrix.groupby("MIGPUMA")["filler"].sum(), left_on="MIGPUMA", right_index=True, how="left")
```

```python
migration_matrix.to_csv("ky_grav_data.csv")
```
